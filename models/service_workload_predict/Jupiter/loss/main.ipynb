```python
import matplotlib.pyplot as plt

models = [ 'LSTM', 'Bi-LSTM', 'Bi-LSTM-Attention', 'N-Beats']
mae_losses = [0.126775, 0.145264, 0.093982, 0.129477]
mape_losses = [ 33.792211, 39.735156, 25.441057, 34.384256]
mse_losses = [0.026012, 0.032437, 0.012269,0.026951]
rmse_losses = [ 0.161281, 0.180103, 0.111413, 0.164168]

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Function to add value labels
def add_labels(ax, data):
    for i in range(len(data)):
        ax.text(i, data[i], "{:.6f}".format(data[i]), ha='center', va='bottom')

# Plotting MAE Validation Loss
axs[0, 0].bar(models, mae_losses)
axs[0, 0].set_title('MAE Validation Loss')
add_labels(axs[0, 0], mae_losses)

# Plotting MAPE Validation Loss
axs[0, 1].bar(models, mape_losses)
axs[0, 1].set_title('MAPE Validation Loss')
add_labels(axs[0, 1], mape_losses)


# Plotting MSE Validation Loss
axs[1, 0].bar(models, mse_losses)
axs[1, 0].set_title('MSE Validation Loss')
add_labels(axs[1, 0], mse_losses)


# Plotting RMSE Validation Loss
axs[1, 1].bar(models, rmse_losses)
axs[1, 1].set_title('RMSE Validation Loss')
add_labels(axs[1, 1], rmse_losses)

plt.show()
```
