```python
# Summary
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 示例数据，您需要用实际损失值替换这些值
window_sizes = [2, 3, 5, 8, 10]
loss_values_lstm = [0.012218551914432468, 0.0071203957036647305, 0.0029731219696406386, 0.006094271086510233, 0.0070847464105186705]
loss_values_bi_lstm = [0.006471874841551668, 0.0039197325416585735, 0.0029905655231567537, 0.003600082360080822, 0.00418700505588533]
loss_values_attention = [0.009569015747637313, 0.008352652195818799, 0.006823701983896798,0.003794369264844803, 0.004775501182077451]
loss_values_n_beats = [0.0041367343192530975, 0.004436705337752232, 0.003592793756529171, 0.00563317957211373, 0.004056491574159579]

# 创建一个 Pandas DataFrame
data = pd.DataFrame({'Time Window Size': window_sizes * 4,
                     'Loss Value': loss_values_lstm + loss_values_bi_lstm + loss_values_attention + loss_values_n_beats,
                     'Model': ['LSTM'] * len(window_sizes) + ['Bi-LSTM'] * len(window_sizes) + ['Bi-LSTM-Attention'] * len(window_sizes) + ['N-Beats'] * len(window_sizes)})

# 绘制带有置信区间的折线图
sns.lineplot(data=data, x='Time Window Size', y='Loss Value', hue='Model', marker='o')

# 显示图形
plt.show()
```
